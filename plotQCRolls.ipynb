```python

def plotQCRolls(rownum, prh, rolls):
    i = rownum
    sI = rolls.loc[i, 'sFEI']
    eI = rolls.loc[i, 'eFEI']
    rolldat = pd.DataFrame(prh['roll'][sI:eI + 1])
    acceldat = pd.DataFrame(prh['Aw'][sI:eI + 1,])
    depthdat = pd.DataFrame(prh['p'][sI:eI+1])
    pdat = pd.concat([rolldat, acceldat, depthdat], axis=1)
    pdat.columns = ['roll', 'accel_x', 'accel_y', 'accel_z', 'depth']
    
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(18,12), sharex=True)
    
    ax1.clear()
    ax1.plot(pdat['roll'], label='Roll (rad)',  linewidth=2)
    ax1.set_title('Expert audit, roll#: '+str(i), fontsize=20)
    
    sFPI = rolls.loc[i, 'sFPI']- sI
    eFPI = rolls.loc[i, 'eFPI']- sI
    ax1.axvline(sFPI, color = 'black', linestyle = '--')
    ax1.axvline(eFPI, color = 'black', linestyle = '--')
            
    if rolls.loc[i, 'ObsInd'] is not None:
        odat = rolls.loc[[i],['Behavior', 'Comment','ObsInd']]
        odat = odat.explode(['Behavior', 'Comment','ObsInd'])
        odat = odat.reset_index(drop=True)
        
        colors = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
                 for i in range(len(odat))]
        
        for ii in range(len(odat)):
            obsX = odat.loc[ii, 'ObsInd'] - rolls.loc[i, 'sFEI']
            ax1.axvline(obsX, color = colors[ii])
            ax1.annotate(odat.loc[ii, 'Behavior'], xy=(obsX, 0),rotation = 90)
    else:
        print('Obs not Noted')

    ax2.clear()
    ax2.plot(pdat['accel_x'], label='X', linewidth=2)
    ax2.plot(pdat['accel_y'], label='Y',  linewidth=2)
    ax2.plot(pdat['accel_z'], label='Z',  linewidth=2)
    ax2.set_title('Accelerometer')
    ax2.legend()
    
    ax3.clear()
    ax3.plot(pdat['depth'], label='depth(m)',  linewidth=2)                           
    ax3.set_title('Depth (m)')
    ax3.set(ylim=[0, max(pdat['depth'])+5])
    ax3.invert_yaxis()
    
    # Button to add row to 'toRemove' list
    def add_to_remove_list(b):
        toRemove.append(rownum)
        print(f"Row number {rownum} added to 'toRemove' list.")
    
    button = widgets.Button(description="Remove from feeding events")
    button.on_click(add_to_remove_list)
    display(button)
    
    plt.show()
```
